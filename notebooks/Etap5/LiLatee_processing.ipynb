### Uzupełnienie reszty danych

```python
import os
import json

import google_auth_oauthlib.flow
import googleapiclient.discovery
import googleapiclient.errors

jeden = None
def fillInfos(df, step=50):
    # 50 is max results count returned by YT API
    global jeden
    for i in range(0, df.shape[0], step):
        print(f'FilledNoe {i}', end='\r')
        ids_list = df[i:i+step]["video_id"].values.tolist()

        try:
            result = searchVideosByListOfIds(ids_list)
        except Exception as err:
            print("except")
            print(err)
            continue
        
        jeden = result['items']
        for item in result['items']:
            if 'snippet' in item:
                if 'categoryId' in item['snippet']:
                    df.loc[df.video_id==item['id'], ["category_id"]] = item['snippet']['categoryId']
                if 'tags' in item['snippet']:
                    df.loc[df.video_id==item['id'], ["tags"]] = json.dumps(item['snippet']['tags'])
                if 'thumbnails' in item['snippet']:
                    df.loc[df.video_id==item['id'], ["thumbnail_link"]] = item['snippet']['thumbnails']['default']['url']
            if 'statistics' in item:
                if 'viewCount' in item['statistics']:
                    df.loc[df.video_id==item['id'], ["views"]] = item['statistics']['viewCount']
                if 'likeCount' in item['statistics']:
                    df.loc[df.video_id==item['id'], ["likes"]] = item['statistics']['likeCount']
                if 'dislikeCount' in item['statistics']:
                    df.loc[df.video_id==item['id'], ["dislikes"]] = item['statistics']['dislikeCount']
                if 'commentCount' in item['statistics']:
                    df.loc[df.video_id==item['id'], ["comment_count"]] = item['statistics']['commentCount']
#             try:          
#                 df.loc[df.video_id==item['id'], ["category_id"]] = item['snippet']['categoryId']
#                 df.loc[df.video_id==item['id'], ["tags"]] = json.dumps(item['snippet']['tags'])
#                 df.loc[df.video_id==item['id'], ["thumbnail_link"]] = item['snippet']['thumbnails']['default']['url']
                
#                 df.loc[df.video_id==item['id'], ["views"]] = item['statistics']['viewCount']
#                 df.loc[df.video_id==item['id'], ["likes"]] = item['statistics']['likeCount']
#                 df.loc[df.video_id==item['id'], ["dislikes"]] = item['statistics']['dislikeCount']
#                 df.loc[df.video_id==item['id'], ["comment_count"]] = item['statistics']['commentCount']
#             except KeyError as err:
#                 print('KeyError: ', err, ' dla ', item['id'])
#                 df.loc[df.video_id==item['id'], ["category_id"]] = ''
#                 df.loc[df.video_id==item['id'], ["tags"]] = ''
#                 df.loc[df.video_id==item['id'], ["thumbnail_link"]] = ''
                
#                 df.loc[df.video_id==item['id'], ["views"]] = ''
#                 df.loc[df.video_id==item['id'], ["likes"]] = ''
#                 df.loc[df.video_id==item['id'], ["dislikes"]] = ''
#                 df.loc[df.video_id==item['id'], ["comment_count"]] = ''
#                 continue
        
    return df

def searchVideosByListOfIds(ids_list):
    api_key = os.environ.get('YT_API')

    # Disable OAuthlib's HTTPS verification when running locally.
    # *DO NOT* leave this option enabled in production.
    os.environ["OAUTHLIB_INSECURE_TRANSPORT"] = "1"

    youtube = googleapiclient.discovery.build(
        "youtube", "v3", developerKey=api_key)

    request = youtube.videos().list(
        part=["snippet", 'statistics'],
        maxResults=50,
        id=ids_list,
    )
    response = request.execute()
    return response
```
