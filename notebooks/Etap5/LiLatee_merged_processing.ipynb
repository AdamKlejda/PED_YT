## TF-IDF?

```python
df_nontren = pd.read_csv("../../our_data/Etap5/Etap2/df_merged.csv")
df_tren = pd.read_csv("../../our_data/Etap4/checkpoints/GB_US_objects_str-V2.csv")
del df_tren['category_id']
df_tren = df_tren.rename(columns={'category_id_GT': 'category_id'})

df_tren['trending'] = 1
df_nontren['trending'] = 0

df = pd.concat([df_nontren, df_tren])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

df.loc[df.title_clean.isna(), "title_clean"] = ''
# title_all_words = df.loc[~df.title_clean.isna(), 'title_clean'].values
temp = df.loc[~df.title_clean.isna(), ['trending','title_clean']]
temp = temp.loc[temp.trending != -1, :]
cat_titles = temp.groupby(by=['trending'])['title_clean'].apply(lambda x: ' '.join(x)).reset_index()
corpus = cat_titles['title_clean'].values

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)

unique_cats_ids = temp.trending.unique()
cnt = 0
def TFIDF_of_column(row, column_name):
    global cnt
    text = row[column_name]
    
    for cat in range(len(unique_cats_ids)):
        sum_of_tfidf = 0
        for word in text.split():
            try:     
                sum_of_tfidf += X.toarray()[cat, vectorizer.get_feature_names().index(word)]
            except ValueError:
                continue
        row[f"TFIDF_{column_name}_{int(unique_cats_ids[cat])}"] = sum_of_tfidf
        
    if cnt % 100 == 0:
        print(cnt, end='\r')
    
    cnt+=1
    return row

column = "title_clean"
df = df.apply(lambda row: TFIDF_of_column(row, column), axis=1)
df.to_csv(f"./../../our_data/Etap5/Etap3/TFIDF_title_clean.csv",index=False)
```
