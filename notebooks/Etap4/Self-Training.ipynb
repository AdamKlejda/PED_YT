```python
def prepareSets(df, chosen_columns, verbose=False):
    X_labeled = df.loc[df.category_id != -1, chosen_columns]
    y_labeled = df.loc[df.category_id != -1, ['category_id_GT']].values.ravel()

    X_unlabeled = df.loc[df.category_id == -1, chosen_columns]
    y_unlabeled = df.loc[df.category_id == -1, ['category_id', 'category_id_GT']]


    X_mixed = np.concatenate((X_labeled, X_unlabeled))
    y_mixed = np.concatenate((y_labeled, y_unlabeled['category_id']))
    
    if verbose:
        print(f'X_labeled: ', X_labeled.shape)
        print(f'y_labeled: ', y_labeled.shape)
        print(f'X_unlabeled: ', X_unlabeled.shape)
        print(f'y_unlabeled: ', y_unlabeled.shape)
        print(f'X_mixed: ', X_mixed.shape)
        print(f'y_mixed: ', y_mixed.shape)
    
    return X_labeled, y_labeled, X_unlabeled, y_unlabeled, X_mixed, y_mixed

```

# Self Learning

```python
import numpy as np
from sklearn import datasets
from sklearn.semi_supervised import SelfTrainingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import itertools as it
from sklearn.pipeline import Pipeline
from sklearn.model_selection import ShuffleSplit, StratifiedKFold
from sklearn.metrics import accuracy_score
import warnings
import json

warnings.filterwarnings("ignore")
```

```python
k_best_columns = 20

sets_of_columns = {
    'r1': chosen_columns_r1[:k_best_columns], 
    'r2': chosen_columns_r2[:k_best_columns], 
    'r3': chosen_columns_r3[:k_best_columns],
    'all': chosen_columns_r1
}

results = {}

kernel = [{'kernel': 'poly'}, {'kernel': 'rbf'}, {'kernel': 'sigmoid'}] 
gamma = [{'gamma':'auto'},{'gamma':'scale'}]
n_neighbors = [{'n_neighbors': 3}, {'n_neighbors': 5}, {'n_neighbors': 7}, {'n_neighbors': 9}]
# gamma = [{'gamma': 1.0},  {'gamma': 10.0}]
# alpha = [{'alpha': 0.1}, {'alpha': 0.2}, {'alpha': 0.4}]
solver = [{'solver':'newton-cg'},{'solver':'lbfgs'},{'solver':'sag'},{'solver':'saga'}]
C = [{'C':0.01},{'C':0.1},{'C':1},{'C':10},{'C':100}]
algorithm = [{'algorithm':'ball_tree'},{'algorithm':'kd_tree'},{'algorithm':'brute'}]
weights = [{'weights':'uniform'},{'weights':'distance'}]

LR_params_temp = list(it.product(solver, C))
SVC_params_temp = list(it.product(kernel, C,gamma))
KNN_params_temp = list(it.product(n_neighbors,weights,algorithm))

LR_params = []
for params in LR_params_temp:
    temp_dict = {}
    for param in params:
        temp_dict[list(param.keys())[0]] = list(param.values())[0]
    LR_params.append(temp_dict)
     
SVC_params = []
for params in SVC_params_temp:
    temp_dict = {}
    for param in params:
        temp_dict[list(param.keys())[0]] = list(param.values())[0]
    SVC_params.append(temp_dict)
    
KNN_params = []
for params in KNN_params_temp:
    temp_dict = {}
    for param in params:
        temp_dict[list(param.keys())[0]] = list(param.values())[0]
    KNN_params.append(temp_dict)
        

    
sets_results_LR = {}
sets_results_SVC = {}
sets_results_KNN = {}

for set_name, chosen_columns in sets_of_columns.items():
    X_labeled, y_labeled, X_unlabeled, y_unlabeled, X_mixed, y_mixed = prepareSets(df, chosen_columns, verbose=False)
    
    for params in LR_params:
        LR = LogisticRegression(n_jobs=-1, 
                                   solver=params['solver'], 
                                   C=params['C'])
        modelLR = SelfTrainingClassifier(LR)
        modelLR.fit(X_mixed, y_mixed)
        yhat = modelLR.predict(X_unlabeled)

        score = accuracy_score(y_unlabeled['category_id_GT'], yhat)
 
        print('Accuracy: %.3f' % (score*100))
        print(params)
        
        if set_name in sets_results_LR:
            sets_results_LR[set_name].append((score, params))
        else:
            sets_results_LR[set_name] = []
            sets_results_LR[set_name].append((score, params))
            
    for params in SVC_params:
        svcmod = SVC(probability=True,
                                 kernel=params['kernel'], 
                                 gamma=params['gamma'], 
                                 C=params['C'])
        print(params)
        modelSVC = SelfTrainingClassifier(svcmod)
        modelSVC.fit(X_mixed, y_mixed)
        yhat = modelSVC.predict(X_unlabeled)

        score = accuracy_score(y_unlabeled['category_id_GT'], yhat)
 
        print('Accuracy: %.3f' % (score*100))
        print(params)
        
        if set_name in sets_results_SVC:
            sets_results_SVC[set_name].append((score, params))
        else:
            sets_results_SVC[set_name] = []
            sets_results_SVC[set_name].append((score, params))
        
    for params in KNN_params:
        KNNmod = KNeighborsClassifier(n_jobs=-1, 
                                 algorithm=params['algorithm'], 
                                 weights=params['weights'], 
                                 n_neighbors=params['n_neighbors'])
        
        modelKNN = SelfTrainingClassifier(KNNmod)
        modelKNN.fit(X_mixed, y_mixed)
        yhat = modelKNN.predict(X_unlabeled)

        score = accuracy_score(y_unlabeled['category_id_GT'], yhat)
 
        print('Accuracy: %.3f' % (score*100))
        print(params)
        
        if set_name in sets_results_KNN:
            sets_results_KNN[set_name].append((score, params))
        else:
            sets_results_KNN[set_name] = []
            sets_results_KNN[set_name].append((score, params))
        
# with open(f'../../our_data/Etap4/results/LR.json', 'w') as file:
#     file.write(json.dumps(sets_results_LR))
    
# with open(f'../../our_data/Etap4/results/SVC.json', 'w') as file:
#     file.write(json.dumps(sets_results_SVC))

# with open(f'../../our_data/Etap4/results/KNN.json', 'w') as file:
#     file.write(json.dumps(sets_results_KNN))
```

```python
def checkIfExists(all_params, check_param):

    for acc, p in all_params:
        if p.keys() == check_param[1].keys():
            same = True
            for key in p.keys():
                if p[key] != check_param[1][key]:
                    same = False
            if same:
                return True
    return False

def clean(sets_results):
    result = {}
    
    for set_name in sets_results.keys():
        one_set = sets_results[set_name]
        proper_params_list = []
        for res in one_set:
            temp = res[1].copy()
#             if res[1]['kernel'] == 'knn':
#                 temp.pop('gamma', None)
#             elif res[1]['kernel'] == 'rbf':
#                 temp.pop('n_neighbors', None)

            proper_params_list.append((res[0], temp))

            without_duplication = []
            for el in proper_params_list:
                if not checkIfExists(without_duplication, el):
                    without_duplication.append(el)
        
        result[set_name] =  without_duplication
        
    return result
k_best_columns = 20
sets_of_columns = {
    'r1': chosen_columns_r1[:k_best_columns], 
    'r2': chosen_columns_r2[:k_best_columns], 
    'r3': chosen_columns_r3[:k_best_columns],
    'all': chosen_columns_r1
}

with open(f'../../our_data/Etap4/results/LR.json', 'r') as file:
    sets_results_LR_clean = json.loads(file.read())

with open(f'../../our_data/Etap4/results/SVC.json', 'r') as file:
    sets_results_SVC_clean = json.loads(file.read())
    
with open(f'../../our_data/Etap4/results/KNN.json', 'r') as file:
    sets_results_KNN_clean = json.loads(file.read())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


def change_width(ax, new_value) :
    for patch in ax.patches :
        current_width = patch.get_width()
        diff = current_width - new_value

        # we change the bar width
        patch.set_width(new_value)

        # we recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
```
