```python
import pandas as pd
```

### Dla Etapu 3

```python
br_et3 = pd.read_csv("../our_data/Etap3/GB-Etap3.csv")
us_et3 = pd.read_csv("../our_data/Etap3/US-Etap3.csv")
dfGb= pd.read_csv("../our_data/Etap2/adam/dfGB_finall.csv")
dfUS= pd.read_csv("../our_data/Etap2/adam/dfUS_finall.csv")

dfGb1= pd.read_csv("../our_data/Etap1/GB.csv")
dfUS1= pd.read_csv("../our_data/Etap1/US.csv")
```

```python
# br_et3
dfGb_finall = br_et3.merge(dfGb, how='outer', left_on='video_id', right_on='id')
dfUS_finall = us_et3.merge(dfUS, how='outer', left_on='video_id', right_on='id')
# dfGb_finall = dfGb_finall.merge(dfGb1, how='inner', left_on='video_id', right_on='video_id')
# dfUS_finall = dfUS_finall.merge(dfUS1, how='inner', left_on='video_id', right_on='video_id')
listaGb = dfGb_finall.loc[:, "video_id"].values
listaUS = dfUS_finall.loc[:, "video_id"].values


df_gb1 = dfGb1.loc[~dfGb1.video_id.isin(listaGb), :]
df_us1 = dfUS1.loc[~dfUS1.video_id.isin(listaUS), :]

dfGb_finall = pd.concat([df_gb1,dfGb_finall])
dfUS_finall = pd.concat([df_us1,dfUS_finall])


dfUS_finall = dfUS_finall.drop(['id'], axis=1)
dfUS_finall = dfUS_finall.loc[~dfUS_finall.video_id.isna(), :]
dfUS_finall = dfUS_finall.groupby(['video_id']).first()
dfUS_finall = dfUS_finall.loc[:, dfUS_finall.columns != "Unnamed: 0"]
dfUS_finall = dfUS_finall.reset_index()

dfGb_finall = dfGb_finall.drop(['id'], axis=1)
dfGb_finall = dfGb_finall.loc[~dfGb_finall.video_id.isna(), :]
dfGb_finall = dfGb_finall.groupby(['video_id']).first()
dfGb_finall = dfGb_finall.loc[:, dfGb_finall.columns != "Unnamed: 0"]
dfGb_finall = dfGb_finall.reset_index()
# print(result.shape) 
dfGb_finall.to_csv("../our_data/Etap3/GB_merged.csv",index=False)

dfUS_finall.to_csv("../our_data/Etap3/US_merged.csv",index=False)
```

```python
print("ETAP1  ",dfGb1.shape)
print("ETAP2 moje ",dfGb.shape)
print("ETAP3      ",br_et3.shape)
print("połączone  ",dfGb_finall.shape)
print()
print("ETAP1  ",dfUS1.shape)
print("ETAP2 moje ",dfUS.shape)
print("ETAP3      ",us_et3.shape)
print("połączone  ",dfUS_finall.shape)
```
